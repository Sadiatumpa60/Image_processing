#Edge detection

```python
import numpy as np
import matplotlib.pyplot as plt

def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')

def plot_comparison(img_original, img_filtered, img_title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 8), sharex=True, sharey=True)
    ax1.imshow(img_original, cmap=plt.cm.gray)
    ax1.set_title('Original')
    ax1.axis('off')
    ax2.imshow(img_filtered, cmap=plt.cm.gray)
    ax2.set_title(img_title_filtered)
    ax2.axis('off')
```

```python
from skimage import color
from skimage.filters import sobel

soaps_image = plt.imread('/content/drive/MyDrive/Image processing/Ala_Idris (100).png')

# Make the image grayscale
soaps_image_gray = color.rgb2gray(color.rgba2rgb(soaps_image))

# apply edge detection filters
edge_sobel = sobel(soaps_image_gray)

# Show Comparison
plot_comparison(soaps_image, edge_sobel, "Edge with Sobel")
```

```python
# Import module and function
from skimage.filters import sobel
from skimage import color ,data
# Load the rocket image
rocket= data.rocket()
# Apply edge detection filter
edge_sobel = sobel(rocket)

# Show original and resulting image to compare
plot_comparison(rocket, edge_sobel, "Edge with Sobel")
```

#Blurring to reduce noise

```python
from skimage.filters import gaussian

# Read the image
building_image = plt.imread('/content/drive/MyDrive/Image processing/Ala_Idris (100).png')

# Apply filter
gaussian_image = gaussian(building_image, multichannel=True)

# Show Comparison
plot_comparison(building_image, gaussian_image, "Gaussian Smoothing")
```

#Medical images

```python
from skimage import exposure

blood = plt.imread('/content/drive/MyDrive/Image processing/BloodImage_00001.jpg')

# Show original x-ray image and its histogram
show_image(blood, 'Blood Cell')
```

```python
plt.title('Histogram of image')
plt.hist(blood.ravel(), bins=256);
```

```python
xray_image_eq = exposure.equalize_hist(blood)

# Show the resulting image
show_image(xray_image_eq, 'Resulting image')
```

#Resizing

```python
from skimage.transform import resize

image= plt.imread('/content/drive/MyDrive/Image processing/images.jpg')

# Height and width to resize
height = 400
width = 500

# Resize image
image_resized = resize(image, (height, width), anti_aliasing=True)

# Show the original and resulting images
show_image(image, 'Original image')
show_image(image_resized, 'Resized image')
```
